```python

# Imports
import numpy as np
import pandas as pd 
from scipy import stats

# Leitura do arquivo .csv por meio da biblioteca pandas
students_df = pd.read_csv("predict_students_dropout_and_academic_success.csv", sep = ";")
students_df.columns = students_df.columns.str.strip()
```

```python

# Drop da coluna alvo para manipulação do dataframe
y = students_df["Target"];
students_df.drop("Target", axis = 1, errors = "ignore", inplace = True)
```

```python
# 1. Medidas de Tendência Central
def medidas_tendencia_central(students_df, column):
    series = students_df[column]
    media = series.mean()
    mediana = series.median()  # Funciona para qualquer tamanho de DF!
    ponto_medio = (series.max() + series.min()) / 2
    moda = series.mode()[0]

    return {
        "Media": media,
        "Mediana": mediana,
        "Ponto Medio": ponto_medio,
        "Moda": moda,
    }


# 2. Medidas de Dispersão
def medidas_dispersao(students_df, column):
    series = students_df[column]
    amplitude = series.max() - series.min()
    variancia = series.var(ddof=0)
    desvio_padrao = series.std(ddof=0)
    coef_variacao = (desvio_padrao / media) * 100 if "media" in locals() else (desvio_padrao / series.mean()) * 100

    return {
        "Amplitude": amplitude,
        "Variancia": variancia,
        "Desvio Padrao": desvio_padrao,
        "Coef. Variacao": coef_variacao,
    }


# 3. Medidas de Posição Relativa
def medidas_posicao_relativa(students_df, column):
    series = students_df[column]
    score_z = stats.zscore(series)
    quartil1 = series.quantile(0.25)
    quartil2 = series.quantile(0.50)
    quartil3 = series.quantile(0.75)
    amplitude_quartil = quartil3 - quartil1

    return {
        "Menor Score-Z": score_z.min(),
        "Maior Score-Z": score_z.max(),
        "Quartil 1": quartil1,
        "Quartil 2": quartil2,
        "Quartil 3": quartil3,
        "Amplitude Interquartil": amplitude_quartil,
    }


# 4. Medidas de Associação
def medidas_associacao(students_df, column1, column2):
    covariancia = np.cov(students_df[column1], students_df[column2], ddof=0)[
        0, 1
    ]
    correlacao = students_df[column1].corr(students_df[column2])
    print(f"Covariância entre {column1} e {column2}: {covariancia:.4f}")
    print(f"Correlação entre {column1} e {column2}: {correlacao:.4f}")
```
