```python
# Imports
import pandas as pd

# Leitura do Dataset
students_df = pd.read_csv("predict_students_dropout_and_academic_success.csv", sep = ";")
students_df.columns = students_df.columns.str.strip()
```

<p style="text-align: justify; font-size: 1.15em;">
A importação da biblioteca <strong>Pandas</strong> é crucial para a Análise Exploratória realizada no processo de Mineração de Dados, pois viabiliza a manipulação da base de dados e a leitura de arquivos no formato <code>.csv</code>, conforme executado na primeira célula de código. A função <code>pandas.read_csv()</code> é a responsável por carregar o arquivo estruturado em linhas e colunas, gerando o <em>DataFrame</em> com os dados que serão utilizados durante toda a pesquisa. Por fim, o comando <code>students_df.columns.str.strip()</code> atua diretamente na limpeza dos <strong>nomes das colunas</strong>, removendo espaços em branco errôneos e quebras de linha desnecessárias nos cabeçalhos, o que previne falhas de indexação durante a manipulação dos atributos.
</p>

<p style="text-align: justify; font-size: 1.15em;">
A função <code>.info()</code> é responsável por gerar as informações sobre cada coluna da base de dados, informando a quantidade de colunas, o tipo de dado de cada coluna, seus respectivos nomes e contagem de valores não nulos. Assim, é perceptível que dos <strong>36 atributos</strong>, tem-se <strong>uma variável</strong> alvo do tipo string, ou seja, uma tarefa de classificação com múltiplos rótulos.
```python
# Verificação Desbalanceamento Rótulos da Coluna Target
students_df['Target'].value_counts()
```

<p style="text-align: justify; font-size: 1.15em;">
A verificação do <strong>balanceamento da classe Target</strong> se fez possível com a função <code>.value_counts()</code>, que revela a existência de 3 rótulos, Graduate, Dropout e Enrolled e mostra uma disparidade de amostras rotuladas entre eles. Concluiu-se que para a tarefa de classificação realizada posteriormnete se faz necessária a utilização de técnicas de superamostragem, como <strong>SMOTE</strong>, e de subamostragem, <strong>Undersampling</strong>, para que se tenha um comparativo direto entre o resultado dos experimentos dos modelos preditivos e como a classe majoritária influencia na predição das classes minoritárias.
